```python
import pandas as pd
import numpy as np

# 1. Load the dataset
file_path = '/content/customer_shopping_behavior.csv'
df = pd.read_csv(file_path)

print("Initial shape:", df.shape)

# 2. Handle missing values based on logic
# If Online/Offline is 'Offline', the Online Store and Delivery Speed will be missing.
df['Online Store'] = df['Online Store'].fillna('N/A - Offline')
df['Delivery Speed'] = df['Delivery Speed'].fillna('N/A - Offline')

# Missing values in Festival/Sale mean it was a regular shopping day
df['Festival/Sale'] = df['Festival/Sale'].fillna('Regular Day')

# Missing sizes generally represent accessories or items without standard sizing
df['Size'] = df['Size'].fillna('Not Applicable')

# 3. Standardize and Format Data Types
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

# 4. Feature Engineering for better SQL/Tableau analysis
# Create Age Group Bins
age_bins = [0, 25, 40, 60, 100]
age_labels = ['Gen Z (18-25)', 'Millennials (26-40)', 'Gen X (41-60)', 'Senior']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=True)

# Calculate Total Value (Price * Quantity) and Net Revenue (including Shipping Charge)
df['Total Purchase Amount (₹)'] = df['Purchase Amount (₹)'] * df['Quantity']
df['Net Revenue (₹)'] = df['Total Purchase Amount (₹)'] + df['Shipping Charge (₹)']

# 5. Clean column names for easier SQL querying (remove spaces and special characters)
df.columns = df.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('₹', '').str.replace('/', '_').str.replace('%', 'Pct')

print("Cleaned shape:", df.shape)
print("Missing values remaining:\n", df.isnull().sum())

# 6. Save the cleaned dataset
df.to_csv('cleaned_customer_shopping_behavior.csv', index=False)
print("Cleaned dataset saved as 'cleaned_customer_shopping_behavior.csv'")
```
